```python
import subprocess
import time
import os

import matplotlib.pyplot as plt
```

```python
def run_matrix_multiplication(exe_path, size, path_a, path_b, path_c):
    args = [exe_path, str(size), path_a, path_b, path_c]
    start_time = time.perf_counter()
    result = subprocess.run(args, capture_output=True, text=True)
    end_time = time.perf_counter()
    return end_time - start_time
```

```python
times = []
sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
for i in sizes:
        path_a = os.path.join("..", "gen_matrix", str(i), "a_matrix.txt")
        path_b = os.path.join("..", "gen_matrix", str(i), "b_matrix.txt")
        path_c = os.path.join("results", f"{i}.txt")
        exe_path = os.path.join("mul.exe")
        elapsed = run_matrix_multiplication(exe_path, i, path_a, path_b, path_c)
        times.append(elapsed)
```

```python
times_ms = [t * 1000 for t in times]

plt.figure(figsize=(10, 6))
plt.plot(sizes, times_ms, marker='o')
plt.xlabel("Размер матрицы (NxN)")
plt.ylabel("Время выполнения (мс)")
plt.title("Зависимость времени от размера матриц")
plt.grid(True)
plt.tight_layout()
plt.show()
```
